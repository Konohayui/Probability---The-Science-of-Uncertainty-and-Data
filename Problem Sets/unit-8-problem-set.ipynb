## Convergence in Probability
#### Problem 1
For each of the following sequences, determine whether it converges in probability to a constant. 

1. Let $X_1, X_2, \dots$ be independent continuous random variables, each uniformly distributed between $-1$ and $1$. 
    * Let $U_i=\frac{X_1+X_2+\cdots+X_i}{i}, i=1, 2,\dots$. What value does the sequence $U_i$ converge to in probability?
    
    The sequence $\{U_i\}$ converges to $0$. For all $\epsilon>0$, we have
    \begin{align*}\lim_{i\to\infty}\mathbb{P}(|U_i-0|\geq\epsilon)&=\lim_{i\to\infty}\mathbb{P}\left(\left|\frac{X_1+X_2+\cdots+X_i}{i}-0\right|\geq\epsilon\right)\\
    &=\lim_{i\to\infty}\mathbb{P}\left(\left|X_1+X_2+\cdots+X_i\right|\geq\epsilon\cdot i\right)\\
    &\leq\lim_{i\to\infty}\frac{E[|X_1+X_2+\cdots+X_i|]}{\epsilon\cdot i}\\
    &\leq\lim_{i\to\infty}\frac{E[|X_1|+|X_2|+\cdots+|X_i|]}{\epsilon\cdot i}\\
    &=\lim_{i\to\infty}\frac{E[|X_1|]+E[|X_2|]+\cdots+E[|X_i|]}{\epsilon\cdot i}\\
    &=\lim_{i\to\infty}\frac{1}{2^i\cdot\epsilon\cdot i}\\
    &=0
    \end{align*}
    
    * Let $\sum_i=X_1+X_2+\cdots+X_i, i=1, 2,\dots$. What value does the sequence $\sum_i$ converge to in probability?
    
    The sequence $\{\sum_i\}$ does not converge. Let $\sum_n=X_1+X_2+\cdots+X_n$ and suppose $\sum_{n-1}$ converges to $c$ in probability. Then we have $X_n=\sum_{n}-\sum_{n-1}=0$ which contraditcts $X_i$ is $U[-1,1]$ for all $i$.
    
    * Let $I_i=1$ if $X_i\geq1/2$, and $I_i=0$, otherwise. Define,
    $$S_i=\frac{I_1+I_2+\cdots+I_i}{i}.$$ 
    What value does the sequence $S_i$ converge to in probability?
    
    Sequence $\{S_i\}$ converges to $1/4$. Note that $\mathbb{P}(I_i=1)=\mathbb{P}(X_i\geq1/2)=1/4$. For all $\epsilon>0$, we have 
    \begin{align*}
    \lim_{i\to\infty}\mathbb{P}(|S_i-1/4|\geq\epsilon)&=\lim_{i\to\infty}\mathbb{P}\left(\left|\frac{I_1+I_2+\cdots+I_i}{i}-\frac{1}{4}\right|\geq\epsilon\right)\\
    &\leq\lim_{i\to\infty}\mathsf{Var}\left(\frac{I_1+I_2+\cdots+I_i}{i}\right)\Big/\epsilon^2\\
    &=\lim_{i\to\infty}\frac{\mathsf{Var}(I_1+I_2+\cdots+I_i)}{i^2\epsilon^2}\\
    &=\lim_{i\to\infty}\frac{\mathsf{Var}(I_1)+\mathsf{Var}(I_2)+\cdots+\mathsf{Var}(I_i)}{i^2\epsilon^2}\\
    &=\lim_{i\to\infty}\left(\frac{1}{4}\cdot\frac{3}{4}\right)^i\frac{1}{i^2\epsilon^2}\\
    &=0
    \end{align*}
    
    * Let $W_i=\max\{X_1,X_2,\cdots,X_i\},i=1, 2,\dots,$. What value does the sequence $W_i$ converge to in probability?
    
    The given sequence converges to $1$. Since $-1\leq W_i\leq 1$, we have $|W_i-1|\leq 2$. For $\epsilon>2$, we trivally have $$\lim_{i\to\infty}\mathbb{P}(|W_i-1|\geq\epsilon)=0$$
    Suppose $\epsilon\in(0,2]$, we have
    \begin{align*}
    \lim_{i\to\infty}\mathbb{P}(|W_i-1|\geq\epsilon)&=\lim_{i\to\infty}\mathbb{P}(1-W_i\geq\epsilon)\\
    &=\lim_{i\to\infty}\mathbb{P}(W_i\leq1-\epsilon)\\
    &=\lim_{i\to\infty}\mathbb{P}(\max\{X_1,X_2,\cdots,X_i\}\leq1-\epsilon)\\
    &=\lim_{i\to\infty}\mathbb{P}(X_1\leq1-\epsilon)\cdots\mathbb{P}(X_i\leq1-\epsilon)\\
    &=\lim_{i\to\infty}\left(1-\frac{\epsilon}{2}\right)^i\\
    &=0
    \end{align*}
    
    * Let $V_i=X_1\cdot X_2\cdots X_i,i=1,2,\dots$. What value does the sequence $V_i$ converge to in probability?
    
    The sequence converges to $0$. Note that $|X_i|\leq1$ for all $i$, and so 
    $$|V_i|=|X_1||X_2|\cdots|X_i|\leq\min\{|X_1|,|X_2|,\dots,|X_i|\}\leq1$$
    Hence, for any $\epsilon>1$, we hace $\lim_{i\to\infty}\mathbb{P}(|V_i-0|\geq\epsilon)=0$.
    
    For $\epsilon\in(0,1]$, we have
    \begin{align*}
    \lim_{i\to\infty}\mathbb{P}(|V_i-0|\geq\epsilon)&=\lim_{i\to\infty}\mathbb{P}(|X_1\cdot X_2\cdots X_i-0|\geq\epsilon)\\
    &=\lim_{i\to\infty}\mathbb{P}(|X_1|\cdot|X_2|\cdots|X_i|\geq\epsilon)\\
    &\leq\lim_{i\to\infty}\mathbb{P}(\min\{|X_1|,|X_2|,\dots,|X_i|\}\geq\epsilon)\\
    &=\lim_{i\to\infty}\mathbb{P}(|X_1|\geq\epsilon)\mathbb{P}(|X_2|\geq\epsilon)\cdots\mathbb{P}(|X_i|\geq\epsilon)\\
    &=\lim_{i\to\infty}(1-\epsilon)^i\\
    &=0
    \end{align*}
    
2.  Let $X_1,X_2,\dots,$ be independent identically distributed random variables with $E[X_i]=2$ and $\mathsf{Var}(X_i)=9$, and let $Y_i=X_i/2^i$.
    * What value does the sequence $Y_i$ converge in probability?
    
    The sequence converges to $0$. We have $E[Y_i]=E[X_i]/2^i=1/2^{i-1}$ and $\mathsf{Var}(Y_i)=\mathsf{Var}(X_i)/2^{2i}=9/2^{2i}$. By the Chebyshev inequality, for any $\epsilon>0$,
    $$\mathbb{P}\left(\left|Y_i-\frac{1}{2^{i-1}}\right|\geq\epsilon\right)\leq\frac{9}{2^{2i}\cdot\epsilon^2}$$
    Then, we have

    * Let $A_n=\frac{1}{n}\sum_{i=1}^{n}Y_i$. What value does the sequence $A_n$ converge to in probability?

    \begin{align*}

    &=\frac{1}{n}\sum_{i=1}^{n}\frac{2}{2^i}\\
    &=\frac{1}{n}\left(2-\frac{2}{2^n}\right)
    \end{align*}
    \begin{align*}
    \mathsf{Var}(A_n)&=\mathsf{Var}\left(\frac{1}{n}\sum_{i=1}^{n}Y_i\right)\\
    &=\frac{1}{n^2}\mathsf{Var}\left(\sum_{i=1}^{n}\frac{X_i}{2^i}\right)\\
    &=\frac{1}{n^2}\left(\sum_{i=1}^{n}\frac{\mathsf{Var}(X_i)}{2^{2i}}\right)\\
    &=\frac{1}{n^2}\left(3-\frac{3}{2^{2n}}\right)
    \end{align*}
    
    So, $\lim_{n\to\infty}E[A_n]=0$ and $\lim_{n\to\infty}\mathsf{Var}(A_n)=0$.
    
    By the Chebyshev inequality, for any $\epsilon>0$,
    $$\mathbb{P}\left(\left|A_n-\frac{1}{n}\left(2-\frac{2}{2^n}\right)\right|\geq\epsilon\right)\leq\frac{1}{n^2\epsilon^2}\left(3-\frac{3}{2^{2n}}\right)$$
    Taking the limit as $n\to\infty$, we have
    $$\lim_{n\to\infty}\mathbb{P}(|A_n-0|\geq\epsilon)=0$$
    
    * Let $Z_i=\frac{1}{3}X_i+\frac{2}{3}X_{i+1}$ for $i=1, 2,\dots$, and let $M_n=\frac{1}{n}\sum_{i=1}^{n}Z_i$ for $n=1, 2,\dots$. What value does the sequence $M_n$ converge to in probability?
    
    We have

    By the Weal Law of Large Numbers, the first term converges to $2/3$ and the second term converges to $4/3$ respectively in probability. Therefore, $\{M_n\}$ converges to $2$ in probability.

```python
def X_i(n):
    return np.random.uniform(-1, 1, n)

def plot_sequence(seq, n, ax, title = ""):
    sns.distplot(seq, bins = 100, hist = True, ax = ax)
    ax.set_title("Sequence {} with n = {}".format(title, n))
    
n = 10000
N = 1000
    
fig, ax = plt.subplots(3, 2, figsize = (15, 25))

# U_n = (X_1+X_2+...+X_n)/n
U_n = np.sum([X_i(n) for _ in range(N)], axis = 0)/n

# sum_n = X_1+X_2+...+X_n
sum_n = np.sum([X_i(n) for _ in range(N)], axis = 0)

# I_i = 1 if X_i >= 1/2 else 0
# S_n = (I_1+I_2+...+I_n)/n
S_n = np.sum([np.array(X_i(n) >= 0.5, dtype = int) for _ in range(N)], axis = 0)/n

# W_n = max{X_1,X_2,...,X_n}
W_n = np.max([X_i(n) for _ in range(N)], axis = 0)

# V_n = V_1*V_2*...*V_n
V_n = np.prod([X_i(n) for _ in range(N)], axis = 0)
sequences = {"$U_n$":U_n, "$\sum_n$":sum_n, "$S_n$":S_n, "$W_n$":W_n, "$V_n$":V_n}

for i, seq in enumerate(sequences):
    plot_sequence(sequences[seq], N, ax[i//2][i%2], title = seq)

plt.show()
```

Let $S_n$ be the number of successes in $n$ independent Bernoulli trials, where the probability of success at each trial is $1/3$.

Note that $S_n=X_1+X_2+\cdots+X_n$, where the $X_i$ are independent Bernoulli random variables with parameter $1/3$. Then, $E[S_n]=n/3$ and $\mathsf{Var}(S_n)=2n/9$.

1. $$\lim_{n\to\infty}\mathbb{P}\left(\frac{n}{3}-10\leq S_n\leq\frac{n}{3}+10\right)$$

For any $\epsilon>0$, for sufficuently large $n$, we have $\epsilon\sqrt{n}>10$
\begin{align*}
\lim_{n\to\infty}\mathbb{P}\left(\frac{n}{3}-10\leq S_n\leq\frac{n}{3}+10\right)&\leq\lim_{n\to\infty}\mathbb{P}\left(\frac{n}{3}-\epsilon\sqrt{n}\leq S_n\leq\frac{n}{3}+\epsilon\sqrt{n}\right)\\
&=\lim_{n\to\infty}\mathbb{P}\left(-\epsilon\sqrt{n}\leq S_n-\frac{n}{3}\leq\epsilon\sqrt{n}\right)\\
&=\lim_{n\to\infty}\mathbb{P}\left(-\frac{\epsilon\sqrt{n}}{\sqrt{2n/9}}\leq\frac{S_n-n/3}{\sqrt{2n/9}}\leq\frac{\epsilon\sqrt{n}}{\sqrt{2n/9}}\right)\\
&=\lim_{n\to\infty}\mathbb{P}\left(-\frac{3}{\sqrt{2}}\epsilon\leq\frac{S_n-n/3}{\sqrt{2n/9}}\leq\frac{3}{\sqrt{2}}\epsilon\right)\\
&=\Phi\left(\frac{3}{\sqrt{2}}\epsilon\right)-\Phi\left(-\frac{3}{\sqrt{2}}\epsilon\right)
\end{align*}
Since this is true for every $\epsilon>0$, this remains true as $\epsilon$ approaches to $0$.
$$\lim_{\epsilon\to0}\left[\Phi\left(\frac{3}{\sqrt{2}}\epsilon\right)-\Phi\left(-\frac{3}{\sqrt{2}}\epsilon\right)\right]=\Phi(0)-\Phi(0)=0$$

2. $$\lim_{n\to\infty}\mathbb{P}\left(\frac{n}{3}-\frac{n}{6}\leq S_n\leq\frac{n}{3}+\frac{n}{6}\right)$$

Note that $\frac{n}{3}-\frac{n}{6}\leq S_n\leq\frac{n}{3}+\frac{n}{6}=\left|\frac{S_n}{n}-\frac{1}{3}\right|\leq\frac{1}{6}$, and $E[S_n/n]=1/3$.

For any $\epsilon>1/6$, we simply have $\lim_{n\to\infty}\mathbb{P}\left(\left|\frac{S_n}{n}-\frac{1}{3}\right|\leq\epsilon\right)=0$

For $\epsilon\leq1/6$, we have
\begin{align*}
\lim_{n\to\infty}\mathbb{P}\left(\left|\frac{S_n}{n}-\frac{1}{3}\right|\leq\epsilon\right)
&\leq\lim_{n\to\infty}\frac{\mathsf{Var}(S_n/n)}{\epsilon^2}\\
&=\lim_{n\to\infty}\frac{2}{9}\frac{n}{n^2\epsilon^2}\\
&=0
\end{align*}
The probability of the event converges to $1$.

3. $$\lim_{n\to\infty}\mathbb{P}\left(\frac{n}{3}-\frac{\sqrt{2n}}{5}\leq S_n\leq\frac{n}{3}+\frac{\sqrt{2n}}{5}\right)$$

## The Sample Mean
#### Problem 3
Let $X$ be a continuous random variable. We know that it takes values between $0$ and $6$, but we do not know its distribution or its mean and variance, although we know that its variance is at most $4$. We are interested in estimating the mean of $X$, which we denote by $h$. To estimate $h$, we take $n$ i.i.d. samples $X_1,\dots,X_n$, which all have the same distribution as $X$, and compute the sample mean $$H=\frac{1}{n}\sum_{i=1}^{n}X_i$$

1. Find $E[H]$ and the smallest upper bound for $\mathsf{Var}(H)$.

$$E[H]=E\left[\frac{1}{n}\sum_{i=1}^{n}X_i\right]=\frac{1}{n}E[X_1+X_2+\cdots+X_n]=\frac{1}{n}\times\,nh=h$$
$$\mathsf{Var}(H)=\mathsf{Var}\left(\frac{1}{n}\sum_{i=1}^{n}X_i\right)=\frac{1}{n^2}\mathsf{Var}(X_1+X_2+\cdots+X_n)\leq\frac{1}{n^2}\times\,4n=\frac{4}{n}$$

2. Calculate the smallest possible value of $n$ such that the standard deviation of $H$ is guaranteed to be at most $0.01$.

$$\sqrt{\frac{4}{n}}=\frac{2}{\sqrt{n}}\leq0.01\Rightarrow\,n\geq40000$$

3. We would like to be at least $96\%$ sure that our estimate is within $0.02$ of the true mean $h$. Using the Chebyshev inequality, calculate the minimum value of $n$ that will achieve this.

Apply the Chebyshev inequality to $H$, we have 
$$\mathbb{P}(|H-h|\geq\,0.02)\leq\,\frac{\sigma_H^2}{0.02^2}$$
Apply the upper bound we obtained from part 2, we have
$$\frac{\sigma_H^2}{0.02^2}\leq\frac{4}{n\cdot\,0.02^2}\leq\,1-0.96\Rightarrow\,n\geq\,250000$$

4. Suppose now that $X$ is uniformly distributed on $[h-3, h+3]$, for some unknown $h$. Using the Central Limit Theorem, identify the most appropriate expression for a $95\%$ confidence interval for $h$.

With the Central Limit Theorem, we have
$$\mathbb{P}\left(\left|\frac{H-h}{\sigma_h/\sqrt{n}}\right|\leq1.96\right)\approx\,0.95$$
So,
$$\mathbb{P}\left(H-\frac{1.96\sqrt{3}}{\sqrt{n}}\leq\,h\leq\,H+\frac{1.96\sqrt{3}}{\sqrt{n}}\right)\approx\,0.95$$
Thus, the $95\%$ confidence interval for $h$ is 
$$\left[H-\frac{1.96\sqrt{3}}{\sqrt{n}},H+\frac{1.96\sqrt{3}}{\sqrt{n}}\right]$$

## Airline Overbooking
#### Problem 4
For any given flight, an airline tries to sell as many tickets as possible. Suppose that on average, $20\%$ of ticket holders fail to show up, all independent of one another. Knowing this, an airline will sell more tickets than there are seats available (i.e., overbook the flight) and hope that there is a sufficient number of ticket holders who do not show up, to compensate for its overbooking. Using the Central Limit Theorem, determine $n$, the maximum number of tickets an airline can sell on a flight with 400 seats so that it can be approximately $99\%$ confident that all ticket holders who do show up will be able to board the plane. Use the de Moivre-Laplace  $1/2$-correction in your calculations.

Let $N$ be the total number of ticketed passengers that show up. Then, $N$ has a binomial distribution with parameters, $n$ and $p=0.8$. Therefore,
$$E[N]=0.8n, \mathsf{Var}(N)=n*0.2*0.8=0.16n$$

To find $n$, we need to compute $\mathbb{P}(N>400)$ as the probability that the airline will have deny passengers from boardings is the probability that more than 400 passengers show up.
\begin{align*}
\mathbb{P}(N>400)&=\mathbb{P}(N\geq\,400+0.5)\\
&=\mathbb{P}\left(\frac{N-0.8n}{\sqrt{0.16n}}\geq\frac{400.5-0.8n}{\sqrt{0.16n}}\right)\\
&\approx\,1-\Phi\left(\frac{400.5-0.8n}{\sqrt{0.16n}}\right)
\end{align*}

To obtain $99\%$ confident, we need
$$\Phi\left(\frac{400.5-0.8n}{\sqrt{0.16n}}\right)\approx\,0.99$$
which implies
$$\frac{400.5-0.8n}{\sqrt{0.16n}}=2.33\Rightarrow\,n\approx475$$

```python
n_tickets = [400, 425, 450, 475, 500]
fig, ax = plt.subplots(2, 3, figsize = (25, 15))
for i, tickets in enumerate(n_tickets):
    ticket_holders = np.random.binomial(tickets, 0.8, size = 10000)
    sns.distplot(ticket_holders, bins = 100, hist = True, ax = ax[i//3][i%3])
    ax[i//3][i%3].set_title("Total number of tickets sell {}".format(tickets))
plt.show()
```

## Maximum Likelihood Estimation
#### Problem 5
The random variables $X_1,X_2,\dots,X_n$ are continuous, independent, and distributed according to the Erlang PDF
$$f_X(x)=\frac{\lambda^3x^2e^{-\lambda\,x}}{2},\,\,\,\text{for $x\geq\,0$}$$
where $\lambda$ is an unknown parameter. Find the maximum likelihood estimate of $\lambda$, based on observed values $x_1,x_2,\dots,x_n$.

With observations $x_1,x_2,\dots,x_n$, we receive
$$f_X(x;\,\lambda)=\frac{\lambda^3x_1^2e^{-\lambda\,x}}{2}\cdots\,\frac{\lambda^3x_n^2e^{-\lambda\,x}}{2}$$
To maximize it, we do the followings
$$\ln\left(\frac{\lambda^3x_1^2e^{-\lambda\,x}}{2}\cdots\,\frac{\lambda^3x_n^2e^{-\lambda\,x}}{2}\right)=3n\ln(\lambda)+\ln\left(\prod_{i=1}^{n}x_i^2\right)-\lambda\left(\sum_{i=1}^{n}x_i\right)-n\ln(2)$$
By taking the derivative of the right hand side of the equation above with respect to $\lambda$ and setting it to zero, we have
$$\frac{3n}{\lambda}-\sum_{i=1}^{n}x_i=0\Rightarrow\lambda=\frac{3n}{\sum_{i=1}^{n}x_i}$$
